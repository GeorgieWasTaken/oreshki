### 5) **Чек за покупку**
**Задание:**  
Напишите программу, которая печатает чек за покупку с подробным описанием.

**Формат ввода:**  
1-я строка — название товара.  
2-е число — цена товара.  
3-е число — вес товара.  
4-е число — сумма, внесённая покупателем.

**Формат вывода:**  
```
Чек
<название товара> - <вес>кг - <цена>руб/кг
Итого: <итоговая стоимость>руб
Внесено: <внесённая сумма>руб
Сдача: <сдача>руб
```

---



````python
product = input()
price = int(input())
weight = int(input())
money = int(input())

total = price * weight
change = money - total

print("Чек")
print(f"{product} - {weight}кг - {price}руб/кг")
print(f"Итого: {total}руб")
print(f"Внесено: {money}руб")
print(f"Сдача: {change}руб")
````

### 6) **Конвертация температуры**

### 7) **Форматирование цены нефти**
**Задание:**  
Вам нужно вычислить и красиво вывести стоимость добытой нефти.

**Формат ввода:**  
1-е число — количество баррелей нефти (вещественное).  
2-е число — цена за баррель в долларах (вещественное).

**Формат вывода:**  
«Итоговая стоимость добычи: <стоимость> USD», где стоимость округлена до двух знаков после запятой.

---



````python
bar = float(input())
price = float(input())
total = bar * price
print(f"Итоговая стоимость добычи: {round(total,2)} USD")
````

### 8) **Средняя цена на топливо**
**Задание:**  
Напишите программу для вычисления средней стоимости топлива на трёх заправках.

**Формат ввода:**  
Три строки, каждая из которых содержит цену топлива на одной из заправок.

**Формат вывода:**  
Средняя стоимость топлива, округлённая до двух знаков после запятой.

---



````python
pr1 = float(input())
pr2 = float(input())
pr3 = float(input())

aver = (pr1 + pr2 + pr3) / 3
print(f"{round(aver,2)}")
````
