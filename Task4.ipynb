## 1. Сумма чисел от 1 до N

**Задание:**  
Напишите программу, которая находит сумму всех чисел от 1 до N.

**Формат ввода:**  
Одно целое число N.

**Формат вывода:**  
Сумма всех чисел от 1 до N.


---

```python
n = int(input())
total = 0
for i in range(1, n + 1):
    total += i
print(total)
```

## 2. Факториал числа

**Задание:**  
Напишите программу, которая вычисляет факториал числа N.

**Формат ввода:**  
Одно целое число N.

**Формат вывода:**  
Факториал числа N.

---

```python
n = int(input())
factorial = 1
for i in range(1, n + 1):
    factorial *= i
print(factorial)
```

## 6.  Сравнение двух последовательностей

**Задание:**  
Напишите программу, которая сравнивает получаемые элементы попарно и выводит, какой элемент больше или они равны друг-другу. Если подали "end" выполнение программые заканчивается.

**Формат ввода:**  
Два списка целых чисел одинаковой длины, разделённых пробелом или end.

**Формат вывода:**  
Для каждой пары элементов из двух списков программа выводит одну из строк:
- `первый элемент больше`
- `второй элемент больше`
- `Элементы равны`


---

```python

while True:
  input_message1 = input('Введите элемент 1 ')
  input_message2 = input('Введите элемент 2 ')
  if input_message1 =='end' or input_message2 =='end':
    break
  input_message1=int(input_message1)
  input_message2=int(input_message2)
  if input_message1 >input_message2:
    print("первый элемент больше")
  elif input_message1 <input_message2:
    print("второй элемент больше")
  else:
    print("Элементы равны")
```

## 7. Вложенные циклы: таблица


> Add blockquote


**Задание:**  
Напишите программу, которая выводит квадратную таблицу NxN, заполненную числами от 1 до N^2.

**Формат ввода:**  
Одно целое число N.

**Формат вывода:**  
Таблица NxN с числами от 1 до N^2.
---

```python
n = int(input())
count = 1
for i in range(n):
    for j in range(n):
        print(count, end=' ')
        count += 1
    print()
```

## 8. Сумма всех цифр числа

**Задание:**  
Напишите программу, которая вычисляет сумму всех цифр введённого числа.

**Формат ввода:**  
Одно целое число.

**Формат вывода:**  
Сумма всех цифр числа.
---

```python
num = input()
total = 0
for digit in num:
    if digit.isdigit():
        total += int(digit)
print(total)
```

## 9. Вложенные циклы: прямоугольник из символов

**Задание:**  
Напишите программу, которая выводит прямоугольник размера NxM из символов "*".

**Формат ввода:**  
Два целых числа N и M.

**Формат вывода:**  
Прямоугольник из символов "*" размера NxM.


---

```python
n = int(input())
m = int(input())

for i in range(n):
    for j in range(m):
        print('*', end='')
    print()
```

## 10. Числа Фибоначчи

**Задание:**  
Напишите программу, которая выводит первые N чисел последовательности Фибоначчи.

**Формат ввода:**  
Одно целое число N.

**Формат вывода:**  
Первые N чисел последовательности Фибоначчи, каждое на новой строке.



```python
n = int(input())
a, b = 0, 1
for i in range(n):
    print(a)
    a, b = b, a + b
```
