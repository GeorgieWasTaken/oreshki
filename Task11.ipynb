# Рекурсия, Исключения

## Теория: Рекурсия
**Рекурсия** — это метод программирования, при котором функция вызывает саму себя. Для успешного использования рекурсии необходимо определить:  
1. **Базовый случай** — условие, при котором функция завершает вызовы самой себя.  
2. **Шаг рекурсии** — часть, в которой функция вызывает саму себя с другими параметрами, приближающими её к базовому случаю.

Пример рекурсивной функции:
```python
def fun(n):
    if n == 0:  # Базовый случай
        return 0
    return n ** fun(n - 1)  # Рекурсивный вызов
```

## Теория: Исключения
**Исключения** — это события, возникающие в процессе выполнения программы, которые нарушают её нормальную работу. В Python исключения обрабатываются с помощью блоков try-except.

Блок try содержит код, который может вызвать исключение.
Блок except обрабатывает исключение, если оно произошло.
Дополнительно можно использовать блоки else (если исключения не было) и finally (код, который выполнится всегда).
```python
try:
    x = int(input("Введите число: "))
    print(10 / x)
except ZeroDivisionError:
    print("Деление на ноль!")
except ValueError:
    print("Это не число!")
finally:
    print("Программа завершена.")
```


# 1. Сумма чисел от 1 до n
# Задание:
Напишите рекурсивную функцию, которая вычисляет сумму всех чисел от 1 до n.

# Формат ввода:
Одно целое число \( n \).

# Формат вывода:
Сумма всех чисел от 1 до \( n \).


````python
def sum_rec(n):
    if n <= 0:
        return 0
    return n + sum_rec(n - 1)

n = int(input())
print(sum_rec(n))
````

# 2. Факториал числа
# Задание:
Напишите рекурсивную функцию, которая вычисляет факториал числа.

# Формат ввода:
Одно целое число \( n \).

# Формат вывода:
Факториал числа \( n \).




````python
def fact(n):
    if n == 0 or n == 1:
        return 1
    return n * fact(n - 1)

num = int(input())
print(fact(num))
````

# 3. Числа Фибоначчи
# Задание:
Напишите рекурсивную функцию, которая вычисляет \( n \)-е число Фибоначчи.

# Формат ввода:
Одно целое число \( n \).

# Формат вывода:
\( n \)-е число Фибоначчи.


````python
def f(n):
    if n <= 1:
        return n
    return f(n - 1) + f(n - 2)

n = int(input())
print(f(n))
````

# 4. Проверка палиндрома
# Задание:
Напишите рекурсивную функцию, которая проверяет, является ли строка палиндромом. (РЕКУРСИЕЙ)

# Формат ввода:
Одна строка.

# Формат вывода:
`"Да"`, если строка является палиндромом, `"Нет"` в противном случае.


````python
def palindrome(s):
    if len(s) <= 1:
        return "да"
    if s[0] != s[-1]:
        return "нет"
    return palindrome(s[1:-1])

text = input()
print(palindrome(text))
````

# 5. Деление с обработкой ошибок
# Задание:
Напишите программу, которая запрашивает два числа и делит первое на второе. Обработайте ошибку деления на ноль и неверного ввода.

# Формат ввода:
Два значения: числа или строки.

# Формат вывода:
Результат деления или сообщение об ошибке.

````python
try:
    num1 = float(input())
    num2 = float(input())
    result = num1 / num2
    print(result)
except ZeroDivisionError:
    print("Деление на ноль невозможно")
except ValueError:
    print("Неверный ввод числа")
except Exception as e:
    print("Произошла ошибка:", e)
````

# 6. Подсчёт суммы цифр числа
# Задание:
Напишите рекурсивную функцию, которая подсчитывает сумму цифр числа.

# Формат ввода:
Одно целое число \( n \).

# Формат вывода:
Сумма цифр числа \( n \).


````python
def sum_digits(n):
    n = abs(int(n))
    if n < 10:
        return n
    return (n % 10) + sum_digits(n // 10)

try:
    num = input()
    result = sum_digits(num)
    print(result)
except Exception:
    print("Произошла ошибка")
````

# 7. Перевод числа в двоичную систему
# Задание:
Напишите рекурсивную функцию, которая переводит число из десятичной системы счисления в двоичную.

# Формат ввода:
Одно целое число \( n \).

# Формат вывода:
Строка, представляющая число в двоичной системе.

````python
def binary(n):
    n = int(n)
    if n == 0:
        return "0"
    if n == 1:
        return "1"
    return binary(n // 2) + str(n % 2)

try:
    num = input()
    if int(num) < 0:
        result = "-" + binary(abs(int(num)))
    else:
        result = binary(num)
    print(result)
except Exception:
    print("Произошла ошибка")
````

# 8. Глубина вложенных списков
# Задание:
Напишите рекурсивную функцию, которая вычисляет максимальную глубину вложенности списков.

# Формат ввода:
Список, содержащий элементы или другие списки.

# Формат вывода:
Максимальная глубина вложенности.

````python
import ast

def max_depth(data):
    if not isinstance(data, list):
        return 0
    if not data:
        return 1
    return 1 + max(max_depth(item) for item in data)

try:
    lst = ast.literal_eval(input())
    if not isinstance(lst, list):
        raise ValueError("не список")
    result = max_depth(lst)
    print(result)
except Exception:
    print("ошибка")
````

# 9. Поиск минимального элемента
# Задание:
Напишите рекурсивную функцию, которая находит минимальный элемент в списке.

# Формат ввода:
Список чисел.

# Формат вывода:
Минимальный элемент.

````python
import ast

def recursive_min(lst):
    if not lst:
        raise ValueError("пусто")
    if len(lst) == 1:
        return lst[0]
    else:
        min_rest = recursive_min(lst[1:])
        return lst[0] if lst[0] < min_rest else min_rest

try:
    data = ast.literal_eval(input())
    if not isinstance(data, list):
        raise ValueError("не список")
    if not data:
        raise ValueError("пусто")


    for item in data:
        if not isinstance(item, (int, float)):
            raise ValueError("не числа")
    result = recursive_min(data)
    print(result)
except RecursionError:
    print("превышена максимальная глубина рекурсии")
except ValueError as e:
    print(f"Ошибка: {e}")
except Exception:
    print("ошибка")
````

# 10. Исключения при работе со списками
# Задание:
Напишите программу, которая запрашивает индекс элемента списка и выводит его значение. Обработайте ошибки: индекс за пределами списка и некорректный ввод.

# Формат ввода:
Список из нескольких элементов и индекс.

# Формат вывода:
Элемент списка или сообщение об ошибке.

````python
import ast

try:

    data_input = input().strip()
    lst = ast.literal_eval(data_input)


    if not isinstance(lst, list):
        raise TypeError("не список")

    index_input = input().strip()


    try:
        index = int(index_input)
    except ValueError:
        raise ValueError("не целое")


    if index < 0 or index >= len(lst):
        raise IndexError(f"индекс {index} находится за пределами списка длиной {len(lst)}")

    print(lst[index])

except (SyntaxError, ValueError, TypeError) as e:
    print(f"ошибка ввода: {e}")
except IndexError as e:
    print(f"ошибка индекса: {e}")
except Exception as e:
    print(f"ошибка: {e}")
````
