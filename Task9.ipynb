# Формат ввода:
В функцию подается два числа.
#
# Формат вывода:
Выведите наибольший общий делитель.


```python
def nod(a, b):
    while b != 0:
        a, b = b, a % b
    return a
nod(14,35)
```

# 4.  Имя of the month
Разработайте функцию month, которая принимает номер месяца и обозначение языка ("ru", "en") и возвращает название заданного месяца в заданном языке с заглавной буквы.

Примечание
Ваше решение должно содержать только функции.
В решении не должно быть вызовов требуемых функций.

```python
def month(month_num, lang):
    months_ru = {
        1: 'Январь', 2: 'Февраль', 3: 'Март', 4: 'Апрель', 5: 'Май', 6: 'Июнь',
        7: 'Июль', 8: 'Август', 9: 'Сентябрь', 10: 'Октябрь', 11: 'Ноябрь', 12: 'Декабрь'
    }
    months_en = {
        1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
        7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'
    }

    if lang == 'ru':
        return months_ru.get(month_num, '')
    elif lang == 'en':
        return months_en.get(month_num, '')
    else:
        return ''
month(5,'ru')
```

# 6. Шахматный «обед»
Напишите функцию can_eat, которая принимает положение коня и другой фигуры в виде кортежей из двух координат, а возвращает булево значение: True если конь съедает фигуру и False если нет. В случае фигуры вне поля должно выводиться предупреждение  "Фигура вне поля".

# Формат ввода
Координаты в формате А1, B5
# Формат вывода
True, False или "Фигура вне поля"

```python
def can_eat(horse, figure):
    letters = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7, 'H': 8}

    h_letter, h_num = horse
    f_letter, f_num = figure

    if h_letter not in letters or f_letter not in letters or not (1 <= h_num <= 8) or not (1 <= f_num <= 8):
        return "Фигура вне поля"

    h_x = letters[h_letter]
    h_y = h_num
    f_x = letters[f_letter]
    f_y = f_num

    dx = abs(h_x - f_x)
    dy = abs(h_y - f_y)

    return (dx == 1 and dy == 2) or (dx == 2 and dy == 1)
can_eat(('A',1), ('B',3))
```

# 7. Простая задача
# Задание:
Напишите функцию is_prime, которая принимает натуральное число, а возвращает булево значение: True — если переданное число простое, а иначе — False.

# Формат ввода:
Число
# Формат вывода:
Булевое значение

```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True
is_prime(17)
```

# 8. Слияние
Напишите функцию merge, которая принимает два отсортированных по возрастанию кортежа целых чисел, а возвращает один из всех переданных чисел.

```python
def merge(tuple1, tuple2):
    result = []
    i, j = 0, 0

    while i < len(tuple1) and j < len(tuple2):
        if tuple1[i] < tuple2[j]:
            result.append(tuple1[i])
            i += 1
        else:
            result.append(tuple2[j])
            j += 1

    result.extend(tuple1[i:])
    result.extend(tuple2[j:])

    return tuple(result)
merge([1,2,3],[4,5,6])
```

# 10. Полиндром

Примечание

Для определения типа параметра нужно использовать функцию isinstance.


```python
def is_palindrome(obj):
  return obj == obj[::-1]
is_palindrome('abba')
```
